```python
import pandas as pd

nba_players = pd.read_csv('/content/all_seasons.csv', index_col = 0)

kobe = nba_players[nba_players.player_name == 'Kobe Bryant']
kobe_age = kobe.age.values
kobe_points = kobe.pts.values
seasons = kobe.season.values
```

## Section-II

```python
player_heights = nba_players.groupby('player_name').player_height.mean().values

player_weights =  nba_players.groupby('player_name').player_weight.mean().values

player_pts = nba_players.groupby('player_name').gp.mean().values
```

```python
plt.figure(figsize = (14, 10))
plt.scatter(player_heights, player_weights, alpha=0.7, s= player_pts)
plt.text(155, 65 , 'Muggsy Bogues')
plt.text(226, 165 , 'Sim Bhullar')
plt.text(233, 137 , 'Gheorghe Muresan')
plt.title(" NBA Player's Heights vs Weights ")
plt.xlabel("Player's Height in cm")
plt.ylabel('Players Weight in kg')
plt.style.use('seaborn-poster')
plt.grid()
plt.show()
```
